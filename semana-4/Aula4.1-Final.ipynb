Usaremos dados da UCI Machine Learning Repository
Heart Disease Data Set: https://archive.ics.uci.edu/ml/datasets/Heart+Disease

* 3 age: age in years
* 4 sex: sex (1 = male; 0 = female)
* 9 cp: chest pain type
  * Value 1: typical angina
  * Value 2: atypical angina
  * Value 3: non-anginal pain
  * Value 4: asymptomatic
* 10 trestbps: resting blood pressure (in mm Hg on admission to the hospital)
* 12 chol: serum cholestoral in mg/dl
* 16 fbs: (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
* 19 restecg: resting electrocardiographic results
  * Value 0: normal
  * Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
  * Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
* 32 thalach: maximum heart rate achieved
* 38 exang: exercise induced angina (1 = yes; 0 = no)
* 40 oldpeak = ST depression induced by exercise relative to rest
* 41 slope: the slope of the peak exercise ST segment
  * Value 1: upsloping
  * Value 2: flat
  * Value 3: downsloping
* 44 ca: number of major vessels (0-3) colored by flourosopy
* 51 thal: 3 = normal; 6 = fixed defect; 7 = reversable defect
* 58 num: diagnosis of heart disease (angiographic disease status)
  * Value 0: < 50% diameter narrowing
  * Value 1: > 50% diameter narrowing

`names=['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num']`

```python
df = pd.read_csv('processed.cleveland.data', names=['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num'])
df.describe(include='all')
```

```python
df = df.drop( df.loc[ df['ca']=='?'].index ) 
df = df.drop( df.loc[ df['thal']=='?'].index ) 
df = df.apply( pd.to_numeric )
df.describe(include='all')
```

```python
y = df['num']
df = df.drop(columns=['num'])
print(y)
df.describe()
```

```python
y = y.replace([2,3,4],1)
print(y)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=0, solver='newton-cg')
clf = clf.fit(df,y)
```

```python
pred = clf.predict(df)
print(pred)
```

```python
from sklearn.metrics import *
acc  = accuracy_score(y, pred)
prec = precision_score(y, pred)
rec  = recall_score(y, pred)
print('acc=%.2f, prec=%.2f, rec=%.2f'%(acc,prec,rec))
```

```python
pred_proba = clf.predict_proba(df)
pred_proba = (pred_proba[:,1]>0.3)*1
acc  = accuracy_score(y, pred_proba)
prec = precision_score(y, pred_proba)
rec  = recall_score(y, pred_proba)
print('acc=%.2f, prec=%.2f, rec=%.2f'%(acc,prec,rec))
print(pred_proba)
```

```python
confusion_matrix(y,pred)
```

```python
confusion_matrix(y,pred_proba)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=41)
```

```python

clf = LogisticRegression(random_state=0, solver='newton-cg')
clf = clf.fit(X_train,y_train)
```

```python
pred = clf.predict(X_test)
acc  = accuracy_score(y_test, pred)
prec = precision_score(y_test, pred)
rec  = recall_score(y_test, pred)
print('acc=%.2f, prec=%.2f, rec=%.2f'%(acc,prec,rec))
```

```python
import statsmodels.api as sm

log_reg = sm.Logit(y_train, X_train).fit()
print(log_reg.summary())
```

```python
pred = log_reg.predict(X_test)
pred = (pred>0.5)*1
acc  = accuracy_score(y_test, pred)
prec = precision_score(y_test, pred)
rec  = recall_score(y_test, pred)
print('acc=%.2f, prec=%.2f, rec=%.2f'%(acc,prec,rec))
```
