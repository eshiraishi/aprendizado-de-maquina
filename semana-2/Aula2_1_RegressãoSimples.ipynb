```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
```

```python
df_houses, y_houses = fetch_california_housing(return_X_y=True, as_frame=True)
df_houses.describe()
```

```python
df_houses['prices'] = y_houses
df_houses.corr(method='pearson')
```

```python
X = df_houses['MedInc']
results = sm.OLS(y_houses, sm.add_constant(X)).fit()
print(results.params)
print(results.summary())
plt.plot(X,y_houses,".")
line = np.ones(len(X)) *results.params[0] + X * results.params[1]
plt.plot(X,line)
print(results.rsquared)
print(np.sqrt(results.mse_total))
```

```python
X = df_houses['AveBedrms']
results = sm.OLS(y_houses, sm.add_constant(X)).fit()
print(results.params)
print(results.summary())
plt.plot(X,y_houses,".")
line = np.ones(len(X)) *results.params[0] + X * results.params[1]
plt.plot(X,line)
print(results.rsquared)
print(np.sqrt(results.mse_total))
```

```python
X = np.arange(0,100,0.1)
Y = 10 + 0.1 * X*X + np.random.randn(len(X))*100
results = sm.OLS(Y, sm.add_constant(X)).fit()
print(results.params)
print(results.summary())
plt.plot(X,Y,".")
line = np.ones(len(X)) *results.params[0] + X * results.params[1]
plt.plot(X,line)
print(results.rsquared)
print(np.sqrt(results.mse_total))
```
